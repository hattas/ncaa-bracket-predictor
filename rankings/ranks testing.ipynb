```python
# For loop for each region to get to a single team representing that region in the final four

# basic_stats[basic_stats['School']==]['PPG'].values[0]

for r in region:
    print("-------------"+r)
    
    # Goes through each round until there is a single team
    for tround in round_num:
        
        # prints round number of tourney and teams left in specific region
        #print("-----round " + str(int(tround/2)))
        #print('------------teams left in the %s Region: %s' % (r,str(int(tround/2))))
        
        # goes from 0 to the round number skipping every other
        for i in range(0,tround,2):
            # if the round is 16 then it sets the team names from the original tourney array
            # then after that it will look in the next_round array
            # both are setting the team names for the specific region and rank 
            if tround == 16:
                one_team = ranks['School'][(ranks['Region'] == r) & (ranks["Seed"] == tourney[i])].values[0]
                two_team = ranks['School'][(ranks['Region'] == r) & (ranks["Seed"] == tourney[i+1])].values[0]
            else:
                one_team = ranks['School'][(ranks['Region'] == r) & (ranks["Seed"] == next_round[i])].values[0]
                two_team = ranks['School'][(ranks['Region'] == r) & (ranks["Seed"] == next_round[i+1])].values[0]
            
            # if random int is 0 then the higher ranked team wins otherwise the lower ranked team wins
            # adds the winning team to the next_round advancing through the tourney
            # if there is two teams left then the winner wins that region and advances to the final four
            
            x = basic_stats[basic_stats['School']==one_team]['PPG'].values[0]
            print("one teams ppg: %f" %x)
            y = basic_stats[basic_stats['School']==two_team]['PPG'].values[0]
            print("second teams ppg: %f" %y)
            
            if x>y:
                if tround == 2:
                    final4.append(one_team)
                    print('%s beat %s to advanace to the final four' % (one_team, two_team))
                else:
                    print('%s beat %s to advance to the round of %s' % (one_team, two_team,int(tround*2)))
                    if tround == 16:
                        next_round.append(tourney[i])
                        round32.append(one_team)
                    else:
                        next_round.append(next_round[i])
                        if tround == 8:
                            sweet16.append(one_team)
                        elif tround == 4:
                            elite8.append(one_team)
            else:
                if tround == 2:
                    final4.append(two_team)
                    print('%s beat %s to advance to the final four ' % (two_team, one_team))
                else:
                    print('%s beat %s to advance to the round of %s' % (two_team, one_team,int(tround*2)))
                    if tround == 16:
                        next_round.append(tourney[i+1])
                        round32.append(two_team)
                    else:
                        next_round.append(next_round[i+1])
                        if tround == 8:
                            sweet16.append(two_team)
                        elif tround == 4:
                            elite8.append(two_team)
                        
        # splits the list so that the only elements left are teams still alive
        # after a Region the loop next_round will be empty
        if tround != 16:
            del next_round[0:tround]
            #print(next_round)
            
#print("round of 32: %s" % str((round32)))
#print("round of 16: %s" % str((sweet16)))
#print("round of 8: %s" % str((elite8)))
#print("round of 4: %s" % str((final4)))
print(final4)
```

```python
# Determines the Final game

for i in range(0,2):
 
    x = basic_stats[basic_stats['School']==final4[2*i]]['PPG'].values[0]
    y = basic_stats[basic_stats['School']==final4[2*i+1]]['PPG'].values[0]
    # adds the winning team to the final
    
    # determines losing team based on y 
    if x>y:
        winning_team = final4[2*i]
        finals.append(winning_team)
        losing_team = final4[2*i+1]
    else: 
        winning_team = final4[2*i+1]
        finals.append(winning_team)
        losing_team = final4[2*i]
        
    # prints who makes the finals
    print("%s beat %s to advance to the finals" % (winning_team, losing_team))
```

```python
# Determines Winner of March Madness

# determins random team that will win
x = basic_stats[basic_stats['School']==finals[0]]['PPG'].values[0]
y = basic_stats[basic_stats['School']==finals[1]]['PPG'].values[0]

# sets losing team to opposite 
if x>y:
    winner = finals[0]
    losing_team = finals[1]
else:
    winner = finals[1]
    losing_team = finals[0]

# prints the winner
print("%s beat %s to win the March Madness Tournament" % (winner, losing_team))
```
