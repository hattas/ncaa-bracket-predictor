```python
from random import *
import pandas as pd

# Imports March Madness Rankings from 2015 
ranks = pd.read_csv("1415_ranks.csv", encoding ='latin-1')

# Intializses two arrays that will be used throughout the nested for loops
next_round = []
final_four = []

'''
Order is based on the rankings to get through the tourney 
Round number is based on 16 teams in each region not overall rounds
Declares Regions
'''
tourney = [1,16,8,9,5,12,4,13,6,11,3,14,7,10,2,15]
round_num = [16,8,4,2]
region = ['Midwest','West', 'East', 'South']

# For loop for each region to get to a single team representing that region in the final four
for r in region:
    print("-------------"+r)
    
    # Goes through each round until there is a single team
    for tround in round_num:
        
        # prints round number of tourney and teams left in specific region
        print("-----round " + str(int(tround/2)))
        print('------------teams left in the %s Region: %s' % (r,str(int(tround/2))))
        
        # goes from 0 to the round number skipping every other
        for i in range(0,tround,2):
            # declares random int either 0 or 1
            x = randint(0,1)
            
            # if the round is 16 then it sets the team names from the original tourney array
            # then after that it will look in the next_round array
            # both are setting the team names for the specific region and rank 
            if tround == 16:
                one_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == tourney[i])].values[0])
                two_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == tourney[i+1])].values[0])
            else:
                one_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == next_round[i])].values[0])
                two_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == next_round[i+1])].values[0])
            
            # if random int is 0 then the higher ranked team wins otherwise the lower ranked team wins
            # adds the winning team to the next_round advancing through the tourney
            # if there is two teams left then the winner wins that region and advances to the final four 
            if x == 0:
                if tround == 2:
                    final_four.append(one_team)
                    print('%s beat %s to advanace to the final four' % (one_team, two_team))
                else:
                    print('%s beat %s to advance to the round of %s' % (one_team, two_team,int(tround*2)))
                    if tround == 16:
                        next_round.append(tourney[i])
                    else:
                        next_round.append(next_round[i])
            else:
                if tround == 2:
                    final_four.append(two_team)
                    print('%s beat %s to advance to the final four ' % (two_team, one_team))
                else:
                    print('%s beat %s to advance to the round of %s' % (two_team, one_team,int(tround*2)))
                    if tround == 16:
                        next_round.append(tourney[i+1])
                    else:
                        next_round.append(next_round[i+1])
                        
        # splits the list so that the only elements left are teams still alive
        # after a Region the loop next_round will be empty
        if tround != 16:
            del next_round[0:tround]

# Final Four loop
print("This is the final four: " + str(final_four))
```
