```python
import pandas as pd
ranks = pd.read_csv("1415_ranks.csv", encoding ='latin-1')
tourney = [1,16,8,9,5,12,4,13,6,11,3,14,7,10,2,15]
from random import *
next_round = []
final_four = []
region = ['Midwest','West', 'East', 'South']
round_num = [16,8,4,2]
for r in region:
    print("-------------"+r)
    for tround in round_num:
        print("-----round " + str(tround))
        print('------------teams left in the %s Region: %s' % (r,str(int(tround/2))))
        for i in range(0,tround,2):
            x = randint(0,1)
            if tround == 16:
                one_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == tourney[i])].values[0])
                two_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == tourney[i+1])].values[0])
            else:
                one_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == next_round[i])].values[0])
                two_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == next_round[i+1])].values[0])
            if x == 0:
                if tround == 2:
                    final_four.append(one_team)
                    print('%s beat %s to advanace to the final four' % (one_team, two_team))
                else:
                    print('%s beat %s to advance to the round of %s' % (one_team, two_team,int(tround*2)))
                    if tround == 16:
                        next_round.append(tourney[i])
                    else:
                        next_round.append(next_round[i])
            else:
                if tround == 2:
                    final_four.append(two_team)
                    print('%s beat %s to advance to the final four ' % (two_team, one_team))
                else:
                    print('%s beat %s to advance to the round of %s' % (two_team, one_team,int(tround*2)))
                    if tround == 16:
                        next_round.append(tourney[i+1])
                    else:
                        next_round.append(next_round[i+1])
        if tround != 16:
            del next_round[0:tround]

# Final Four loop
print("This is the final four: " + str(final_four))
```
