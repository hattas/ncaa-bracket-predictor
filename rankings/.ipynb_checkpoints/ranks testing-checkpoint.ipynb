```python
from random import *
import pandas as pd

# Imports March Madness Rankings from 2014-15 
ranks = pd.read_csv("1415_ranks.csv", encoding ='latin-1')

# Intializses arrays that will be used throughout the nested for loops
next_round = []
round32 = []
sweet16 = []
elite8 = []
final4 = []
finals = []
winner = []

'''
Order is based on the rankings to get through the tourney 
Round number is based on 16 teams in each region not overall rounds
Declares Regions
'''
tourney = [1,16,8,9,5,12,4,13,6,11,3,14,7,10,2,15]
round_num = [16,8,4,2]
region = ['Midwest','West', 'East', 'South']
```

```python
outcomes = pd.read_csv("1415_outcomes.csv", encoding = 'latin-1')
outcomes.head()
round32_actual = []
sweet16_actual = []
elite8_actual = []
final4_actual = []
finals_actual = []
winner_actual = []

for i in range(0,32):
    round32_actual.append(str([outcomes.loc[:,'Round32'].values[i]]))
for i in range(0,16):
    sweet16_actual.append(str([outcomes.loc[:,'Round16'].values[i]]))
for i in range(0,8):
    elite8_actual.append(str([outcomes.loc[:,'Round8'].values[i]]))
for i in range(0,4):
    final4_actual.append(str([outcomes.loc[:,'Round4'].values[i]]))
for i in range(0,2):
    finals_actual.append(str([outcomes.loc[:,'Round2'].values[i]]))
for i in range(0,1):
    winner_actual.append(str(outcomes.loc[:,'Round1'].values[i]))
    
final4_actual
```

```python
# For loop for each region to get to a single team representing that region in the final four
for r in region:
    print("-------------"+r)
    
    # Goes through each round until there is a single team
    for tround in round_num:
        
        # prints round number of tourney and teams left in specific region
        print("-----round " + str(int(tround/2)))
        print('------------teams left in the %s Region: %s' % (r,str(int(tround/2))))
        
        # goes from 0 to the round number skipping every other
        for i in range(0,tround,2):
            # declares random int either 0 or 1
            x = 1
            
            # if the round is 16 then it sets the team names from the original tourney array
            # then after that it will look in the next_round array
            # both are setting the team names for the specific region and rank 
            if tround == 16:
                one_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == tourney[i])].values[0])
                two_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == tourney[i+1])].values[0])
            else:
                one_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == next_round[i])].values[0])
                two_team = str(ranks.loc[:,['School']][(ranks['Region'] == r) & (ranks["Seed"] == next_round[i+1])].values[0])
            
            # if random int is 0 then the higher ranked team wins otherwise the lower ranked team wins
            # adds the winning team to the next_round advancing through the tourney
            # if there is two teams left then the winner wins that region and advances to the final four 
            if x == 0:
                if tround == 2:
                    final4.append(one_team)
                    print('%s beat %s to advanace to the final four' % (one_team, two_team))
                else:
                    print('%s beat %s to advance to the round of %s' % (one_team, two_team,int(tround*2)))
                    if tround == 16:
                        next_round.append(tourney[i])
                        round32.append(one_team)
                    else:
                        next_round.append(next_round[i])
                        if tround == 8:
                            sweet16.append(one_team)
                        elif tround == 4:
                            elite8.append(one_team)
            else:
                if tround == 2:
                    final4.append(two_team)
                    print('%s beat %s to advance to the final four ' % (two_team, one_team))
                else:
                    print('%s beat %s to advance to the round of %s' % (two_team, one_team,int(tround*2)))
                    if tround == 16:
                        next_round.append(tourney[i+1])
                        round32.append(two_team)
                    else:
                        next_round.append(next_round[i+1])
                        if tround == 8:
                            sweet16.append(two_team)
                        elif tround == 4:
                            elite8.append(two_team)
                        
        # splits the list so that the only elements left are teams still alive
        # after a Region the loop next_round will be empty
        if tround != 16:
            del next_round[0:tround]
            
#print("round of 32: %s" % str((round32)))
#print("round of 16: %s" % str((sweet16)))
#print("round of 8: %s" % str((elite8)))
#print("round of 4: %s" % str((final4)))
```

```python
# Determines the Final game

for i in range(0,2):
    
    # y can be either 0 or 1 the first iteration
    # and 2 or 3 second iteration
    y = 2*i + randint(0,1)
    
    # adds the winning team to the final
    finals.append(final4[y])
    
    # determines losing team based on y 
    if y%2 == 0:
        losing_team = final4[y+1]
    else: 
        losing_team = final4[y-1]
        
    # prints who makes the finals
    print("%s beat %s to advance to the finals" % (final4[y], losing_team))
```

```python
# Determines Winner of March Madness

# determins random team that will win
z = randint(0,1)

# sets losing team to opposite 
winner = finals[z]
if z == 0:
    losing_team = finals[z+1]
else:
    losing_team = finals[z-1]

# prints the winner
print("%s beat %s to win the March Madness Tournament" % (winner, losing_team))
```

```python
# Comparing our model against the actual outcomes.
games_correct = 0
points = 0

for i in range(0,32):
    if round32_actual[i] == round32[i]:
        points += 1
        games_correct += 1
for i in range(0,16):
    if sweet16_actual[i] == sweet16[i]:
        points += 2
        games_correct += 1
for i in range(0,8):
    if elite8_actual[i] == elite8[i]:
        points += 4 
        games_correct += 1
for i in range(0,4):
    if final4_actual[i] == final4[i]:
        points += 8
        games_correct += 1
for i in range(0,2):
    if finals_actual[i] == finals[i]:
        points += 16
        games_correct += 1
for i in range(0,1):
    if winner_actual[i] == winner[i]:
        points += 32
        games_correct += 1
        
print("Max total points is equal to: %s" % str(32*6))
print("Our Total points is equal to: %s" % str(points))
print("We predicted %s games correct out of %s games!" % (str(games_correct), (32+16+8+4+2+1)))
print("This comes out to be around %s of all games" % str(str(round((games_correct*100)/63))+ "%"))
```
