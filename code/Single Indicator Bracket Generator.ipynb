```python
from random import *
import pandas as pd

# import March Madness Rankings
ranks = pd.read_csv("../csvs/1415/1415_ranks.csv", encoding ='latin-1')
outcomes = pd.read_csv("../csvs/1415/1415_outcomes.csv", encoding = 'latin-1')
basic_stats = pd.read_csv("../csvs/1415/1415_basic.csv", encoding = 'latin-1')

# initialize arrays that will be used in tourny simulation loops
next_round       = []
roundof32        = []
sweet16          = []
elite8           = []
final4           = []
finals           = []
winner           = []
roundof32_actual = []
sweet16_actual   = []
elite8_actual    = []
final4_actual    = []
finals_actual    = []
winner_actual    = []

tourney_order = [1,16,8,9,5,12,4,13,6,11,3,14,7,10,2,15]
regions = ['Midwest','West', 'East', 'South']
```

```python
# For loop for each region to get to a single team representing that region in the final four

# basic_stats[basic_stats['School']==]['PPG'].values[0]

for region in regions:
    print("-------------"+region)
    
    next_round = tourney_order[:] # next round starts off as all teams in region
    
    # Goes through each round until there is a single team
    for round_num in range(1,5):
                
        num_teams = 2**(5-round_num) # number of teams in current round i.e. round 1 has 16 teams per region
        
        # prints round number of tourney and teams left in specific region
        print("-----round " + str(round_num))
        print('------------teams left in the %s Region: %s' % (region,num_teams))
        
        # goes from 0 to the round number skipping every other
        for i in range(0,num_teams,2):
            
            # set the team names for the specific region and rank
            team1 = ranks['School'][(ranks['Region'] == region) & (ranks["Seed"] == next_round[i])].values[0]
            team2 = ranks['School'][(ranks['Region'] == region) & (ranks["Seed"] == next_round[i+1])].values[0]
            
            # get both team's statistics and pick winner/loser
            team1_stat = basic_stats[basic_stats['School']==team1]['OverallSRS'].values[0]
            team2_stat = basic_stats[basic_stats['School']==team2]['OverallSRS'].values[0]
            
            if team1_stat > team2_stat:
                winner = team1
                loser  = team2
                next_round.append(next_round[i])
            else:
                loser  = team1
                winner = team2
                next_round.append(next_round[i+1])

                
            print('%s beat %s to advance to the round of %s' % (winner, loser, num_teams*2))
            if   num_teams == 16:
                roundof32.append(winner)
            elif num_teams == 8:
                sweet16.append(winner)
            elif num_teams == 4:
                elite8.append(winner)
            elif num_teams == 2:
                final4.append(winner)
                        
        # splits the list so that the only elements left are teams still alive
        # after a Region the loop next_round will be empty
        del next_round[0:num_teams]
            
#print("round of 32: %s" % str((roundof32)))
#print("round of 16: %s" % str((sweet16)))
#print("round of 8: %s" % str((elite8)))
#print("round of 4: %s" % str((final4)))
```

```python
# Determines the Final game

for i in range(0,2):
 
    x = basic_stats[basic_stats['School']==final4[2*i]]['OverallSRS'].values[0]
    y = basic_stats[basic_stats['School']==final4[2*i+1]]['OverallSRS'].values[0]
    # adds the winning team to the final
    
    # determines losing team based on y 
    if x>y:
        winning_team = final4[2*i]
        finals.append(winning_team)
        losing_team = final4[2*i+1]
    else: 
        winning_team = final4[2*i+1]
        finals.append(winning_team)
        losing_team = final4[2*i]
        
    # prints who makes the finals
    print("%s beat %s to advance to the finals" % (winning_team, losing_team))
```

```python
# Determines Winner of March Madness

x = basic_stats[basic_stats['School']==finals[0]]['OverallSRS'].values[0]
y = basic_stats[basic_stats['School']==finals[1]]['OverallSRS'].values[0]

# sets losing team to opposite 
if x>y:
    winner = finals[0]
    losing_team = finals[1]
else:
    winner = finals[1]
    losing_team = finals[0]

# prints the winner
print("%s beat %s to win the March Madness Tournament" % (winner, losing_team))
```
