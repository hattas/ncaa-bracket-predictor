```python
from random import *
import pandas as pd

# import March Madness Rankings
ranks = pd.read_csv("../csvs/1415/1415_ranks.csv", encoding ='latin-1')
outcomes = pd.read_csv("../csvs/1415/1415_outcomes.csv", encoding = 'latin-1')
basic_stats = pd.read_csv("../csvs/1415/1415_basic.csv", encoding = 'latin-1')

# initialize arrays that will be used in tourny simulation loops
next_round       = []
roundof32        = []
sweet16          = []
elite8           = []
final4           = []
finals           = []
winner           = []
roundof32_actual = []
sweet16_actual   = []
elite8_actual    = []
final4_actual    = []
finals_actual    = []
winner_actual    = []


best_points = 0

tourney_order = [1,16,8,9,5,12,4,13,6,11,3,14,7,10,2,15]
regions = ['Midwest','West', 'East', 'South']
# For loop for each region to get to a single team representing that region in the final four

# basic_stats[basic_stats['School']==]['PPG'].values[0]

for indicator1 in basic_stats.iloc[:, 2:]:
    for indicator2 in basic_stats.iloc[:, 2:]:
        next_round       = []
        roundof32        = []
        sweet16          = []
        elite8           = []
        final4           = []
        finals           = []
        winner           = []
        roundof32_actual = []
        sweet16_actual   = []
        elite8_actual    = []
        final4_actual    = []
        finals_actual    = []
        winner_actual    = []
        for region in regions:
            #print("-------------"+region)

            next_round = tourney_order[:] # next round starts off as all teams in region

            # Goes through each round until there is a single team
            for round_num in range(1,5):

                num_teams = 2**(5-round_num) # number of teams in current round i.e. round 1 has 16 teams per region

            # prints round number of tourney and teams left in specific region
            #print("-----round " + str(round_num))
            #print('------------teams left in the %s Region: %s' % (region,num_teams))

            # goes from 0 to the round number skipping every other
                for i in range(0,num_teams,2):

                # set the team names for the specific region and rank
                    team1 = ranks['School'][(ranks['Region'] == region) & (ranks["Seed"] == next_round[i])].values[0]
                    team2 = ranks['School'][(ranks['Region'] == region) & (ranks["Seed"] == next_round[i+1])].values[0]

                # get both team's statistics and pick winner/loser
                    team1_stat = 0 
                    team2_stat = 0 
                    q = basic_stats[basic_stats['School']==team1][indicator1].values[0]
                    w = basic_stats[basic_stats['School']==team1][indicator2].values[0]
                    e = basic_stats[basic_stats['School']==team2][indicator1].values[0]
                    r = basic_stats[basic_stats['School']==team2][indicator2].values[0]
            
                    team1_stat = q*w
                    team2_stat = e*r

                    if team1_stat > team2_stat:
                        winner = team1
                        loser  = team2
                        next_round.append(next_round[i])
                    else:
                        loser  = team1
                        winner = team2
                        next_round.append(next_round[i+1])


                #print('%s beat %s to advance to the round of %s' % (winner, loser, num_teams*2))
                    if   num_teams == 16:
                        roundof32.append(winner)
                    elif num_teams == 8:
                        sweet16.append(winner)
                    elif num_teams == 4:
                        elite8.append(winner)
                    elif num_teams == 2:
                        final4.append(winner)

            # splits the list so that the only elements left are teams still alive
            # after a Region the loop next_round will be empty
                del next_round[0:num_teams]

    #print("round of 32: %s" % str((roundof32)))
    #print("round of 16: %s" % str((sweet16)))
    #print("round of 8: %s" % str((elite8)))
    #print("round of 4: %s" % str((final4)))

    # Determines the Final game

        for i in range(0,2):
            x = 0
            y = 0
            g = basic_stats[basic_stats['School']==final4[2*i]][indicator1].values[0]
            h = basic_stats[basic_stats['School']==final4[2*i]][indicator2].values[0]
            j = basic_stats[basic_stats['School']==final4[2*i+1]][indicator1].values[0]
            k = basic_stats[basic_stats['School']==final4[2*i+1]][indicator2].values[0]
            x = g*h
            y = j*k
        # adds the winning team to the final

        # determines losing team based on y 
            if x>y:
                winning_team = final4[2*i]
                finals.append(winning_team)
                losing_team = final4[2*i+1]
            else: 
                winning_team = final4[2*i+1]
                finals.append(winning_team)
                losing_team = final4[2*i]

        # prints who makes the finals
        #print("%s beat %s to advance to the finals" % (winning_team, losing_team))

        # Determines Winner of March Madness
        x=0
        y=0
        a = basic_stats[basic_stats['School']==finals[0]][indicator1].values[0]
        b = basic_stats[basic_stats['School']==finals[0]][indicator2].values[0]
        c = basic_stats[basic_stats['School']==finals[1]][indicator1].values[0]
        d = basic_stats[basic_stats['School']==finals[1]][indicator2].values[0]
        
        x = a*b
        y = c*d

    # sets losing team to opposite 
        if x>y:
            winner = finals[0]
            losing_team = finals[1]
        else:
            winner = finals[1]
            losing_team = finals[0]

    # prints the winner
    #print("%s beat %s to win the March Madness Tournament" % (winner, losing_team))

    # For loops to fill actual outcome arrays
        for i in range(0,32):
            roundof32_actual.append(outcomes['Round32'].values[i])
        for i in range(0,16):
            sweet16_actual.append(outcomes['Round16'].values[i])
        for i in range(0,8):
            elite8_actual.append(outcomes['Round8'].values[i])
        for i in range(0,4):
            final4_actual.append(outcomes['Round4'].values[i])
        for i in range(0,2):
            finals_actual.append(outcomes['Round2'].values[i])
        for i in range(0,1):
            winner_actual.append(outcomes['Round1'].values[i])

    # Comparing our model against the actual outcomes.
        games_correct = 0
        points = 0

        for i in range(0,32):
            if roundof32_actual[i] == roundof32[i]:
                points += 1
                games_correct += 1
        for i in range(0,16):
            if sweet16_actual[i] == sweet16[i]:
                points += 2
                games_correct += 1
        for i in range(0,8):
            if elite8_actual[i] == elite8[i]:
                points += 3 
                games_correct += 1
        for i in range(0,4):
            if final4_actual[i] == final4[i]:
                points += 4
                games_correct += 1
        for i in range(0,2):
            if finals_actual[i] == finals[i]:
                points += 6
                games_correct += 1
        for i in range(0,1):
            if winner_actual[i] == winner[i]:
                points += 8
                games_correct += 1
        #print("Wondering... %s and %s and %s " % (points, indicator1, indicator2))
        if points > best_points: 
            #print("indicators: %s and % s" % (indicator1, indicator2))
            best_points = points
            best_games = games_correct
            best_indicator = indicator1 + " " + indicator2 
        
print("The best indicator is: %s" % best_indicator)
print("Max total points is equal to: %d" % 120)
print("Our Total points is equal to: %d" % best_points)
print("We predicted %d games correct out of %d games!" % (best_games, 63) )
print("This comes out to be about %d%% of all games" % round(best_games*100/63) )
```
