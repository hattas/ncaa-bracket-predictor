```python
import pandas as pd
reg = pd.read_csv("../csvs/kaggle/regular_season_stats.csv", encoding = 'latin-1')
```

```python
reg.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(color_codes=True)
reg2014 = reg[reg['Season'] == 2014]
reg2015 = reg[reg['Season'] == 2015]
reg2016 = reg[reg['Season'] == 2016]
reg2017 = reg[reg['Season'] == 2017]
```

```python
x = reg2014['W']/reg2014['G']
y = reg2014['FG%']/100
fig=sns.regplot(x,y)
fig.set(xlabel='W%',ylabel='FG%',title='FG% vs W%')
```

```python
a = reg2015['W']/reg2015['G']
b = reg2015['FG%']/100
c = reg2016['W']/reg2016['G']
d = reg2016['FG%']/100
e = reg2017['W']/reg2017['G']
f = reg2017['FG%']/100
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4,1, sharey=True, figsize=(10,20))
sns.regplot(x, y, ax=ax1)
sns.regplot(a, b, ax=ax2)
sns.regplot(c, d, ax=ax3)
sns.regplot(e, f, ax=ax4)
```
